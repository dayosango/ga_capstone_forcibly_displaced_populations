**Target variable**: Accommodation Type. 

- I - Individual Accommodation

- Q - Not Individual Accommodation (Other)

```python
#Normalized distribution of the two classes

pd.DataFrame(df['accommodation_type'].value_counts(normalize=True))
```

```python
sns.histplot(df, x='accommodation_type', color='tab:red')
plt.title('Individual Accommodation Type vs Others', pad=20, fontsize=16)
plt.xlabel(None)
plt.savefig('individualaccom_vs_others.png', dpi=300, bbox_inches = "tight")
plt.show();
```

I decided to visualise the distrubution of placements in urban and rural locations. 

Figure shows that most placements are in urban locations.

```python
sns.histplot(df, x='urban_or_rural_location', color='tab:orange')
plt.title('Placements in Urban and Rural Locations', fontsize=16, pad=20)
plt.xlabel(None)
plt.savefig('Urban_vs_rural.png', dpi=300, bbox_inches = "tight")
plt.show();
```

I decided to look at the distribution of I and Q across urban and rural locations.

The figure shows a strong association between urban location and individual accommodation type.

It also indicates that in a rural location, there was a slightly greater likelihood of being allocated a displacement camp than individual accommodation.

```python
#Normalized distributions of the accommodation types across the location types

by_loc_type = df.groupby(['urban_or_rural_location'])['accommodation_type'].value_counts(normalize=True).unstack()
by_loc_type
```

```python
#Stacked bar plot showing the distribution of indivdual and not individual accommodation types 
#across urban and rural locations.

ax = by_loc_type.plot(kind='bar', stacked=True, rot=0)
ax.set_xlabel(None)
ax.set_ylabel('Normalized distributions')
plt.title('Accommodation Type Distribution Across Rural and Urban Locations', pad=20, fontsize=16)
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1))
plt.savefig('accommodation_type_urban_vs_rural.png', dpi=300, bbox_inches = "tight")
plt.show();
```

I decided to look at the distribution of I and Q across the different population types, to investigate which type of population was more likely to be assigned individual accommodation or a displacement camp (not individual accommodation).


The figure shows that *Internally Displaced Populations (IDP)* were **most** likely to be allocated a displacement camp compared to other populations, followed by Returned Internally Displaced Populations (RDP).


Conversely, *Stateless Populations (STP)* were **most** likely to be allocated individual accommodation, closely followed by Returned Refugees (RET).

```python
by_pop_type = df.groupby(['population_type'])['accommodation_type'].value_counts(normalize=True).unstack()
by_pop_type
```

```python
#Stacked bar plot showing the distribution of indivdual and not individual accommodation types 
#across the different population types.

ax = by_pop_type.plot(kind='bar', stacked=True, rot=0)
ax.set_xlabel(None)
ax.set_ylabel('Normalized distributions')
plt.title('Accommodation Type Distribution Across the Different Population Types', pad=20, fontsize=16)
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1))
plt.savefig('accommodation_type_populations.png', dpi=300, bbox_inches = "tight")
plt.show();
```

Furthermore, I explored the accommodation type that was allocated to males and females of the extreme age ranges, infants aged 0-4 years old and adults over 60 years.

The figure indicates that overall, a greater proportion of infants were allocated individual accommodation over dispacmeent camps compared to the proprotion of older people offered individual accommodations.

```python
#Bar plot showing the accommodation type assigned to males and females of the extreme ages

df.groupby(['accommodation_type'])[['female_aged_0-4 years','male_aged_0-4_years', 
                                    'female_aged_over_60_years', 'male_aged_over_60_years']].sum().plot(kind='bar', rot=0, colormap='seismic')

def add_value_label(list_1, list_2):
    for i in range(len(list_1)):
        plt.text(i, list_2[i-1], list_2[i-1], ha='center')
        
def bar_label(values):
    for l in values:
        height = l.get_height()
        
        plt.annotate('{}'.format(height),
                    xy=(l.get_x() + l.get_width() / 2, height),
                    xytext=(0,3),
                    textcoords='offset points',
                    ha='center', va='bottom')

plt.title('Accommodation Type of Males and Females of the Extreme Age Ranges', pad=20, fontsize=16)
# plt.yscale('log')
plt.ylabel('Counts (log)')
plt.legend(loc="upper right", bbox_to_anchor=(1.7, 1))
plt.grid(axis='y', linestyle='--', linewidth=0.3)
plt.savefig('accommodation_type_extreme_ages.png', dpi=300, bbox_inches = "tight")
plt.show();
```

### Age and Gender Correlations

I wanted to visualise any correlations between various age ranges and genders.

I calculated the Pearson correlation coefficients of the age and gender columns, then used a heatmap to visualise the relationships.

The heatmap highlights that overall there are *high* correlations between each of the demogrpahic groups. Specifically: 

* The **greatest** coefficient being **1** - between infant males and females, between 0 and 4 years of age.
* The **lowest** correlation coefficient being **0.73** - between males aged 5-11 years and males over 60 years. 


The high correlation coefficient across the entire breadth of demographics indicates that many of the forcibly displaced are displaced in groups, , for example families, rather than sole individuals.

### Population Type Distribution Across Various Ages and Genders

Firstly, I wanted to visualise the number of individuals of each population type that was reported over the time frame, from 2012 to 2021.

This led to investigating the population type amongst the different demographics to find out more about the backgrounds of the displaced populations.

The x-axis scale was transformed to a logarithmic scale so that it was easier to visualise the differences in numbers of the population types. This was particularly useful as only **293** Returned Internally Dsiplaced Populations were reported but **43248** recorded Refugees (REF).

As shown clearly in the figures, the **greatest** population type was Refugee, closely followed by Asylum Seeker.

The **smallest** population type recorded was Returned Internally Displaced Population.

```python
population_types_sum = df['population_type'].value_counts()
population_types_sum
```

```python
#Population type amongst all of the individuals reported

population_types_sum.plot(kind='barh', logx=True, color='tab:orange')
plt.ylabel('Population Type')
plt.xlabel('Count (log)')
plt.title('Number of Individuals of Each Population Type', fontsize = 16, pad=20)
plt.savefig('populationtype_count.png', dpi=300, bbox_inches = "tight")
plt.show();
```
